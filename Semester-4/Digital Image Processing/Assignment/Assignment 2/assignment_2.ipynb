# Import Library

```python
from PIL import Image, ImageFilter, ImageChops
import numpy as np
```

# Laplacian Filtering

```python
#load image
image = Image.open('image.jpeg').convert("L")
```

```python
image
```

```python
#laplacian filter
laplacian_image = image.filter(ImageFilter.Kernel((3, 3), [0,-1,0,-1,4,-1,0,-1,0], 1, 0))
```

```python
laplacian_image
```

# Unsharp Masking


    f_mask(x,y) = f_image(x,y) - f_blur(x,y)
    f_hb(x,y) = f_image(x,y) + f_mask(x,y) 


```python
#unsharp masking
blur_image = image.filter(ImageFilter.BLUR)

#creating unsharp mask
unsharp_mask = ImageChops.subtract(image,blur_image)

#creating unsharp image
unsharp_image = ImageChops.add(image,blur_image)
```

```python
unsharp_mask
```

```python
unsharp_image
```

# Highboost Filtering


    f_mask(x,y) = f_image(x,y) - f_blur(x,y)
    f_hb(x,y) = f_image(x,y) + f_mask(x,y) * k
    where k = high boost factor , k>1

```python
#HighBoost Filtering
boost_factor = 2

blur_image_1 = image.filter(ImageFilter.BLUR)

highboost_mask = ImageChops.subtract(image,blur_image_1)

highboost_mask = Image.fromarray(np.array(highboost_mask) * boost_factor)

highboost_image = ImageChops.add(image,highboost_mask)
```
