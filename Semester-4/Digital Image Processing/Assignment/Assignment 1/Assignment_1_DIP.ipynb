#Contrast Stretching


##Formula for contrast stretching
##$$pixel_o = (pixel_i-m_i)\times \frac{M_o - m_o}{M_i - m_i} + m_o$$

  ##### pixel_i = intensity of pixel i of input image.
  ##### pixel_o = intensity of pixel i of output image.
  ##### m_i = minimum intensity of input image.
  ##### m_o = minimum intensity of output image.
  ##### M_i = maximum intensity of input image.
  ##### M_o = maximum intensity of output image.

##For Sample.png

```python
pic_array = np.array(pic)
m_i, M_i = min(pic_array.ravel()), max(pic_array.ravel()) #ravel() function will flatten image array
m_o, M_o = 0, 255 

for i in range(0,len(pic_array)):
    for j in range(0,len(pic_array[0])):
        pic_array[i][j] = (pic_array[i][j] - m_i) * (((M_o - m_o)/(M_i - m_i))+m_o)

display(Image.fromarray(pic_array))
```

##For color.png

```python
color_pic_array = np.array(color_pic)
c_m_i, c_M_i = min(color_pic_array.ravel()), max(color_pic_array.ravel())
c_m_o, c_M_o = 0, 255

for i in range(0,len(pic_array)):
    for j in range(0,len(pic_array[0])):
        color_pic_array[i][j] = (color_pic_array[i][j] - c_m_i) * (((c_M_o - m_o)/(c_M_i - c_m_i))+c_m_o)

display(Image.fromarray(color_pic_array))
```

#Image Thresholding

##For Sample.png

```python
pic = pic.convert('L')
pic_array = np.array(pic)
for i in range(len(pic_array)):
    for j in range(len(pic_array[0])):
        if 20 <= pic_array[i][j] <= 40:
            pic_array[i][j] = 0
        else:
            pic_array[i][j] = 255

display(Image.fromarray(pic_array))
```

##For Color.png

```python
# color_pic = color_pic.convert('L')
color_pic_array = np.array(color_pic)
for i in range(len(color_pic_array)):
    for j in range(len(color_pic_array[0])):
        for k in range(len(color_pic_array[0][0])):
          if 20 <= color_pic_array[i][j][k] <= 40:
            color_pic_array[i][j][k] = 0
          else:
            color_pic_array[i][j][k] = 255
display(Image.fromarray(color_pic_array))
```
