##import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#Import the dataset

```python
data = pd.read_csv('Placement_Data.csv')
x = data.iloc[:, :-2].values
y = data.iloc[:, -1].values
```

```python
print(x)
```

#Step-01. Handle missing data

```python
from sklearn.impute import SimpleImputer
imputer_y = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer_y.fit(y.reshape(-1,1))
y = imputer_y.transform(y.reshape(-1,1))
```

```python
print(y)
```

#Step-02. Encoding categorical data that doesn't have binary result

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [2])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

#Step-03. Encoding categorical data that have binary result

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 6] = le.fit_transform(x[:, 6])
```

```python
print(x)
```

#Step-04. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
print(x)
```

#Step-05. Splitting the data into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```
